```python
import pandas as pd
import os

csv_file_path = "./data/titanic_train.csv"
csv_data = pd.read_csv(
    csv_file_path,
    index_col=0,
    usecols=[
        "PassengerId",
        "Survived",
        "Name",
        "Age",
    ],
)

base_filename = "./data/titanic_train_truncated"
file_extension = ".csv"
counter = 1

while os.path.exists(f"{base_filename}_{counter}-{file_extension}"):
    counter += 1

output_file_name = f"{base_filename}_{counter}-{file_extension}"

csv_data.to_csv(output_file_name)
print(f"New file has been saved: {output_file_name}")
csv_data.head()  # test

# output_file_name = "./data/titanic_train_truncated.csv"
# csv_data.to_csv(output_file_name)
# csv_data.head()
```

```python
import pandas as pd
import os

csv_file_path = "./data/titanic_train.csv"
csv_data = pd.read_csv(
    csv_file_path,
    index_col=0,
    header=0,
    usecols=[
        "PassengerId",
        "Survived",
        "Name",
        "Age",
    ],
)

base_filename = "./data/titanic_train_truncated"
file_extension = ".xlsx"
counter = 1
sheet_name = "sheet1"

while os.path.exists(f"{base_filename}_{counter}-{file_extension}"):
    counter += 1

output_file_name = f"{base_filename}_{counter}-{file_extension}"

csv_data.to_excel(output_file_name, sheet_name)
print(f"New file has been saved: {output_file_name}")
csv_data.head()  # test

# output_file_name = "./data/titanic_train_truncated.csv"
# csv_data.to_csv(output_file_name)
# csv_data.head()
```
