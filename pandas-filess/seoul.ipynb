```python
import pandas as pd
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")

data = pd.read_csv("../data/seoul.csv", encoding="euc-kr")
data.shape
```

```python
data["연도"] = pd.to_datetime(data["사용월"], format="%Y%m").dt.year
data["월"] = pd.to_datetime(data["사용월"], format="%Y%m").dt.month
data.head()
```

```python
data["연도"].unique()
data.query("연도 >= 2018").query('호선명 =="2호선"')
```

```python
sorted(data["지하철역"].unique())
```

```python
list(data["지하철역"].str.split("(')"))[1][0]
```

```python
data["지하철역"] = [i[0] for i in data["지하철역"].str.split("(')")]
sorted(data["지하철역"].unique())
```

```python
on_col = [i for i in data.columns if "승차" in i]
on_col
```

```python
data = data[["사용월", "연도", "월", "지하철역"] + on_col]
data.head()
```

```python
data[on_col].sum(axis=1)
data["합계"] = data[on_col].sum(axis=1)
data.head()
```

```python
# 올바른 코드
data_mean = (
    data.groupby("지하철역")[["합계"]]
    .mean()
    .reset_index()
    .rename({"합계": "월평균"}, axis=1)
)
data_mean
```

```python
fig = px.bar(
    data_mean,
    x="지하철역",
    y="월평균",
    title="지하철역별 월 평균 승차인원",
    template="plotly_dark",
)
fig.show()
```
