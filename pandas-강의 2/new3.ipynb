```python
import folium

f = folium.Figure(width=700, height=500)
m = folium.Map(location=[-36.745771448511476, 174.7441187312856], zoom_start=15).add_to(
    f
)
m
```

```python
folium.Marker(
    [-36.745771448511476, 174.7441187312856],
    icon=folium.Icon(color="red", icon="star"),
    tooltip="My Home",
).add_to(m)
m
```

```python
folium.CircleMarker(
    [-36.745771448511476, 174.7441187312856],
    color="black",
    radius=20,
).add_to(m)
m
```
