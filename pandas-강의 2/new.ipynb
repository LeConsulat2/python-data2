```python
import plotly.express as px
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")

df.head()

###df.groupby('column_to_group_by')["column_to_aggregate"].mean()####
groups1 = df.groupby(["island", "sex"])["body_mass_g"].mean().reset_index()
groups1

# Calculate mean of multiple columns
groups2 = (
    df.groupby(["island", "sex"])[["body_mass_g", "bill_length_mm"]]
    .mean()
    .reset_index()
)

sns.barplot(groups2, x="island", y="body_mass_g", hue="sex")
plt.show()

sns.barplot(groups2, x="island", y="bill_length_mm", hue="sex")
plt.show()
```

```python
fig = px.bar(groups1, x="island", y="body_mass_g", color="sex")
fig.show()
```

```python
fig = px.bar(
    groups1,
    x="island",
    y="body_mass_g",
    color="sex",
    text_auto=".0d",
    width=500,
    height=600,
    title="Average Weight by islands",
    barmode="group",
    labels=dict(body_mass_g="kg", island="Island", sex="Sex"),
)
fig.show()
```

```python
# Template

for temp in [
    "ggplot2",
    "seaborn",
    "simple_white",
    "plotly",
    "plotly_white",
    "plotly_dark",
]:
    fig = px.bar(
        groups1,
        x="island",
        y="body_mass_g",
        color="sex",
        barmode="group",
        text_auto=".0d",
        width=700,
        height=500,
        title=f"template: {temp}",
        labels=dict(body_mass_g="Kg", island="", sex="Sex"),
        template=temp,
    )
    fig.show()
```
